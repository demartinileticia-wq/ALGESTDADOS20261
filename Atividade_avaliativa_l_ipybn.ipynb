```python
#1- Classificador de três números
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))
c = int(input("Digite o terceiro número: "))

# Maior
if a >= b and a >= c:
    maior = a
elif b >= a and b >= c:
    maior = b
else:
    maior = c

# Menor
if a <= b and a <= c:
    menor = a
elif b <= a and b <= c:
    menor = b
else:
    menor = c

# Meio
if (a != maior and a != menor) or (a == b or a == c):
    meio = a
elif (b != maior and b != menor) or (b == a or b == c):
    meio = b
else:
    meio = c

print(f"Maior: {maior}, Meio: {meio}, Menor: {menor}")
```

```python
#2- Calculadora de Gorjeta
valor = float(input("Valor da conta: "))
nota = int(input("Nota (1 a 5): "))

if valor <= 0 or nota < 1 or nota > 5:
    print("Entrada inválida")
else:
    if nota == 5:
        gorjeta = valor * 0.20
    elif nota == 4:
        gorjeta = valor * 0.15
    elif nota == 3:
        gorjeta = valor * 0.10
    else:
        gorjeta = 0

    total = valor + gorjeta
    print(f"Gorjeta: R${gorjeta:.2f}")
    print(f"Total: R${total:.2f}")
```

```python
#3- Verificador de Ano bissexto
ano = int(input("Digite o ano: "))

if ano <= 0:
    print("Ano inválido")
else:
    if (ano % 400 == 0):
        print("Bissexto")
    else:
        if (ano % 100 == 0):
            print("Não bissexto")
        else:
            if (ano % 4 == 0):
                print("Bissexto")
            else:
                print("Não bissexto")
```

```python
#4- Cálculo de imposto de renda simplificado
salario = float(input("Salário bruto: "))

if salario <= 0:
    print("Valor inválido")
else:
    if salario <= 2259:
        aliquota = 0
    elif salario <= 2826:
        aliquota = 0.075
    elif salario <= 3751:
        aliquota = 0.15
    elif salario <= 4664:
        aliquota = 0.225
    else:
        aliquota = 0.275

    imposto = salario * aliquota
    liquido = salario - imposto

    print(f"Alíquota: {aliquota*100:.1f}%")
    print(f"Imposto: R${imposto:.2f}")
    print(f"Salário líquido: R${liquido:.2f}")
```

```python
#5- Placar de Futebol
a = int(input("Gols do time A: "))
b = int(input("Gols do time B: "))

if a < 0 or b < 0:
    print("Valores inválidos")
else:
    if a > b:
        vencedor = "Time A"
        diferenca = a - b
    elif b > a:
        vencedor = "Time B"
        diferenca = b - a
    else:
        print("Empate")
        diferenca = 0

    if a != b:
        print(f"Vencedor: {vencedor}")
        print(f"Diferença: {diferenca}")
        if diferenca > 3:
            print("Goleada!")
```

```python
#6- Validador de data
dia = int(input("Dia: "))
mes = int(input("Mês: "))
ano = int(input("Ano: "))

if ano <= 0:
    print("Ano inválido")
elif mes < 1 or mes > 12:
    print("Mês inválido")
else:
    bissexto = False
    if (ano % 400 == 0):
        bissexto = True
    elif (ano % 100 == 0):
        bissexto = False
    elif (ano % 4 == 0):
        bissexto = True

    if mes in (4, 6, 9, 11):
        max_dia = 30
    elif mes == 2:
        if bissexto:
            max_dia = 29
        else:
            max_dia = 28
    else:
        max_dia = 31

    if dia < 1 or dia > max_dia:
        print("Dia inválido")
    else:
        print("Data válida")
```

```python
#7- Calculadora de frete
peso = float(input("Peso (kg): "))
regiao = int(input("Região (1-3): "))

if peso <= 0 or regiao < 1 or regiao > 3:
    print("Entrada inválida")
else:
    if peso <= 1:
        frete = 15
    elif peso <= 5:
        frete = 25
    else:
        frete = 40

    if regiao == 2 or regiao == 3:
        frete *= 1.2

    print(f"Frete: R${frete:.2f}")
```

```python
#8- Simulador de caixa eletrônico
saldo = 1000
senha_correta = 1234

valor = float(input("Valor do saque: "))
senha = int(input("Senha: "))

if valor <= 0:
    print("Valor inválido")
else:
    if senha != senha_correta:
        print("Senha incorreta")
    else:
        if valor > saldo:
            print("Saldo insuficiente")
        else:
            saldo -= valor
            print(f"Saque realizado. Saldo: R${saldo:.2f}")
```

```python
#9- Conversor de temperatura completo
temp = float(input("Temperatura: "))
unidade = input("Unidade (C/F/K): ")

match unidade:
    case "C":
        temp_c = temp
        temp_f = (temp * 9/5) + 32
        temp_k = temp + 273.15
    case "F":
        temp_c = (temp - 32) * 5/9
        temp_f = temp
        temp_k = temp_c + 273.15
    case "K":
        if temp < 0:
            print("Kelvin inválido")
        else:
            temp_c = temp - 273.15
            temp_f = (temp_c * 9/5) + 32
            temp_k = temp
    case _:
        print("Unidade inválida")

print(f"C: {temp_c:.2f}, F: {temp_f:.2f}, K: {temp_k:.2f}")
```

```python
#10- Classificador de IMC com risco
peso = float(input("Peso: "))
altura = float(input("Altura: "))

if peso <= 0 or altura < 0.5 or altura > 2.5:
    print("Valores inválidos")
else:
    imc = peso / (altura ** 2)

    if imc < 18.5:
        classe = "Abaixo do peso"
        risco = "baixo"
    elif imc < 25:
        classe = "Normal"
        risco = "baixo"
    elif imc < 30:
        classe = "Sobrepeso"
        risco = "moderado"
    elif imc < 35:
        classe = "Obesidade I"
        risco = "moderado"
    elif imc < 40:
        classe = "Obesidade II"
        risco = "alto"
    else:
        classe = "Obesidade III"
        risco = "alto"

    print(f"IMC: {imc:.2f}")
    print(f"Classificação: {classe}")
    print(f"Risco: {risco}")
```

```python
#11- Detector de múltiplos
n = int(input("Número: "))

if n <= 0:
    print("Inválido")
else:
    eh2 = (n % 2 == 0)
    eh3 = (n % 3 == 0)
    eh5 = (n % 5 == 0)

    if eh2:
        print("Múltiplo de 2")
    if eh3:
        print("Múltiplo de 3")
    if eh5:
        print("Múltiplo de 5")

    if (eh2 and eh3 and eh5):
        print("Múltiplo de todos")
```

```python
#12- Calculadora de desconto progressivo
preco = float(input("Preço: "))
cupom = input("Cupom: ")

match cupom:
    case "PROMO10":
        desconto = 0.10
    case "PROMO20":
        desconto = 0.20
    case "PROMO30":
        desconto = 0.30
    case "SEMCUPOM":
        desconto = 0
    case _:
        print("Cupom inválido")
        desconto = 0

if preco < 50:
    if desconto > 0.10:
        desconto = 0.10

valor_desc = preco * desconto
final = preco - valor_desc

print(f"Desconto: R${valor_desc:.2f}")
print(f"Final: R${final:.2f}")
```

```python
#13- Avaliador de triângulo completo
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

if a <= 0 or b <= 0 or c <= 0:
    print("Lados inválidos")
else:
    if a < b + c and b < a + c and c < a + b:
        if a == b and b == c:
            print("Equilátero")
        elif a == b or a == c or b == c:
            print("Isósceles")
        else:
            print("Escaleno")
    else:
        print("Não forma triângulo")
```

```python
#14- calculadora de notas com conceito
n1 = float(input("Nota 1: "))
n2 = float(input("Nota 2: "))
n3 = float(input("Nota 3: "))
n4 = float(input("Nota 4: "))

if (0 <= n1 <= 10 and 0 <= n2 <= 10 and
    0 <= n3 <= 10 and 0 <= n4 <= 10):

    media = (n1 + n2 + n3 + n4) / 4

    if media >= 9:
        conceito = "A"
        status = "Aprovado"
    elif media >= 7:
        conceito = "B"
        status = "Aprovado"
    elif media >= 5:
        conceito = "C"
        status = "Recuperação"
    elif media >= 3:
        conceito = "D"
        status = "Reprovado"
    else:
        conceito = "E"
        status = "Reprovado"

    print(f"Média: {media:.2f}")
    print(f"Conceito: {conceito}")
    print(f"Status: {status}")
else:
    print("Notas inválidas")
```

```python
#15- Verificador de acesso por horário
hora = int(input("Digite a hora atual (0-23): "))
tipo = input("Digite o tipo de usuário (admin, operador, visitante): ").lower()

# Validação da hora
if hora < 0 or hora > 23:
    print("Hora inválida!")
else:
    # Determinar turno
    if 6 <= hora <= 11:
        turno = "manhã"
    elif 12 <= hora <= 17:
        turno = "tarde"
    elif 18 <= hora <= 23:
        turno = "noite"
    else:
        turno = "madrugada"

    acesso = "negado"

    match tipo:
        case "admin":
            acesso = "permitido"

        case "operador":
            if turno == "manhã" or turno == "tarde":
                acesso = "permitido"
            else:
                acesso = "negado"

        case "visitante":
            if turno == "tarde":
                acesso = "permitido"
            else:
                acesso = "negado"

        case _:
            print("Tipo de usuário inválido!")
            exit()

    print(f"Acesso {acesso}. Turno atual: {turno}.")
```
