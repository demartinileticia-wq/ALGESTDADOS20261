```python
#15 -
hora = int(input("Digite a hora (0-23): "))
tipo = input("Tipo de usuário (admin/operador/visitante): ").lower()

if hora < 0 or hora > 23:
    print("Hora inválida.")
else:
    # Determinar turno
    if 6 <= hora <= 11:
        turno = "manhã"
    elif 12 <= hora <= 17:
        turno = "tarde"
    elif 18 <= hora <= 23:
        turno = "noite"
    else:
        turno = "madrugada"

    match tipo:
        case "admin":
            acesso = True
        case "operador":
            if turno == "manhã" or turno == "tarde":
                acesso = True
            else:
                acesso = False
        case "visitante":
            if turno == "tarde":
                acesso = True
            else:
                acesso = False
        case _:
            print("Tipo inválido.")
            exit()

    print(f"Turno: {turno}")
    print("Acesso permitido." if acesso else "Acesso negado.")
```

```python
16 -
tipo = input("Tipo de combustível (gasolina/etanol/diesel): ").lower()
litros = float(input("Quantidade de litros: "))

if litros <= 0:
    print("Quantidade inválida.")
else:
    preco_gasolina = 5.80
    preco_etanol = 4.20
    preco_diesel = 6.00

    match tipo:
        case "gasolina":
            preco = preco_gasolina
        case "etanol":
            preco = preco_etanol
        case "diesel":
            preco = preco_diesel
        case _:
            print("Combustível inválido.")
            exit()

    subtotal = litros * preco

    if litros > 40:
        desconto = subtotal * 0.03
    else:
        desconto = 0

    total = subtotal - desconto

    print(f"Subtotal: R${subtotal:.2f}")
    print(f"Desconto: R${desconto:.2f}")
    print(f"Total: R${total:.2f}")
```

```python
17 -
velocidade = float(input("Velocidade (km/h): "))
via = input("Tipo de via (urbana/rodovia/autopista): ").lower()

if velocidade < 0:
    print("Velocidade inválida.")
else:
    match via:
        case "urbana":
            limite = 60
        case "rodovia":
            limite = 100
        case "autopista":
            limite = 120
        case _:
            print("Via inválida.")
            exit()

    if velocidade <= limite:
        situacao = "Dentro do limite"
    elif velocidade <= limite * 1.2:
        situacao = "Excesso leve"
    elif velocidade <= limite * 1.5:
        situacao = "Excesso moderado"
    else:
        situacao = "Excesso grave"

    print(f"Limite: {limite} km/h")
    print(f"Velocidade: {velocidade} km/h")
    print(f"Situação: {situacao}")
```

```python
18 -
valor = float(input("Valor da compra: "))
parcelas = int(input("Número de parcelas (1-12): "))

if valor <= 0 or parcelas < 1 or parcelas > 12:
    print("Dados inválidos.")
else:
    if parcelas == 1:
        juros = 0
    elif parcelas <= 3:
        juros = 0.05
    elif parcelas <= 6:
        juros = 0.10
    else:
        juros = 0.15

    total = valor * (1 + juros)
    valor_parcela = total / parcelas

    print(f"Total com juros: R${total:.2f}")
    print(f"Parcelas: {parcelas}x de R${valor_parcela:.2f}")

    if valor_parcela < 20:
        print("Parcela mínima não atingida.")
        min_parcelas = int(total / 20)
        if total % 20 != 0:
            min_parcelas += 1
        print(f"Mínimo de parcelas: {min_parcelas}")
```

```python
19 -
alvo = float(input("Valor alvo: "))
tolerancia = float(input("Tolerância: "))
medido = float(input("Valor medido: "))

if alvo < 0 or tolerancia < 0 or medido < 0:
    print("Valores inválidos.")
else:
    inferior = alvo - tolerancia
    superior = alvo + tolerancia

    if medido < inferior:
        classificacao = "Abaixo do limite"
    elif medido > superior:
        classificacao = "Acima do limite"
    else:
        classificacao = "Dentro da faixa"

    desvio = medido - alvo

    print(f"Limite inferior: {inferior}")
    print(f"Limite superior: {superior}")
    print(f"Desvio: {desvio}")
    print(f"Classificação: {classificacao}")
```

```python
29 -
atacante = input("Tipo do atacante: ").lower()
defensor = input("Tipo do defensor: ").lower()
poder = float(input("Poder base do ataque: "))

# Validação simples
tipos_validos = ["fogo", "agua", "planta", "eletrico", "terra"]

if atacante not in tipos_validos or defensor not in tipos_validos or poder < 0:
    print("Dados inválidos.")
else:
    vantagem = False
    desvantagem = False

    # Regras de vantagem
    match atacante:
        case "fogo":
            if defensor == "planta":
                vantagem = True
            elif defensor == "agua":
                desvantagem = True

        case "agua":
            if defensor == "fogo":
                vantagem = True
            elif defensor == "planta" or defensor == "eletrico":
                desvantagem = True

        case "planta":
            if defensor == "agua":
                vantagem = True
            elif defensor == "fogo":
                desvantagem = True

        case "eletrico":
            if defensor == "agua":
                vantagem = True
            elif defensor == "terra":
                desvantagem = True

        case "terra":
            if defensor == "eletrico":
                vantagem = True

    # Definir multiplicador
    if vantagem:
        mult = 2
        situacao = "Vantagem"
    elif desvantagem:
        mult = 0.5
        situacao = "Desvantagem"
    else:
        mult = 1
        situacao = "Neutro"

    dano_final = poder * mult

    print(f"Situação: {situacao}")
    print(f"Dano final: {dano_final:.2f}")
```
