```python
# 01 - Levantamento Salarial

soma = 0
maior = 0
acima3000 = 0

for i in range(10):
    salario = float(input("Digite o salário: R$ "))

    soma += salario

    if salario > maior:
        maior = salario

    if salario > 3000:
        acima3000 += 1

media = soma / 10

print("Média salarial:", media)
print("Maior salário:", maior)
print("Funcionários acima de R$3000:", acima3000)
```

```python
# 02 - Desempenho da Turma

soma = 0
qtd = 0
aprovados = 0
reprovados = 0

nota = float(input("Digite a nota (-1 para sair): "))

while nota != -1:
    soma += nota
    qtd += 1

    if nota >= 7:
        aprovados += 1
    else:
        reprovados += 1

    nota = float(input("Digite a nota (-1 para sair): "))

if qtd > 0:
    media = soma / qtd
    print("Média:", media)

print("Aprovados:", aprovados)
print("Reprovados:", reprovados)
```

```python
# 03 - Sistema Bancário de Saques

saldo = float(input("Digite o saldo inicial: R$ "))

saque = float(input("Digite o valor do saque (0 para sair): "))

while saque != 0:

    if saque <= saldo:
        saldo -= saque
        print("Saque realizado!")
    else:
        print("Saldo insuficiente!")

    print("Saldo atual:", saldo)

    saque = float(input("Digite o valor do saque (0 para sair): "))

print("Sistema encerrado.")
```

```python
# 04 - Pesquisa Demográfica

soma = 0
qtd = 0
maior = 0
maiores18 = 0

idade = int(input("Digite a idade (0 para sair): "))

while idade != 0:

    soma += idade
    qtd += 1

    if idade > maior:
        maior = idade

    if idade > 18:
        maiores18 += 1

    idade = int(input("Digite a idade (0 para sair): "))

if qtd > 0:
    media = soma / qtd
    print("Média das idades:", media)

print("Maior idade:", maior)
print("Maiores de 18:", maiores18)
```

```python
# 05 - Controle de Acesso

senha_correta = 2024
tentativas = 0

while tentativas < 3:

    senha = int(input("Digite a senha: "))

    if senha == senha_correta:
        print("Acesso liberado!")
        break
    else:
        print("Senha incorreta!")
        tentativas += 1

if tentativas == 3:
    print("Acesso bloqueado!")
```

```python
# 06 - Vendas com Desconto

total = 0

for i in range(8):

    venda = float(input("Digite o valor da venda: R$ "))

    if venda > 100:
        venda = venda * 0.90

    total += venda

print("Total arrecadado:", total)
```

```python
# 07 - Análise de Sequência Numérica

pares = 0
impares = 0
soma_positivos = 0

num = int(input("Digite um número (0 para sair): "))

while num != 0:

    if num % 2 == 0:
        pares += 1
    else:
        impares += 1

    if num > 0:
        soma_positivos += num

    num = int(input("Digite um número (0 para sair): "))

print("Pares:", pares)
print("Ímpares:", impares)
print("Soma dos positivos:", soma_positivos)
```

```python
# 08 - Monitoramento de Peso

soma = 0
acima80 = 0

peso = float(input("Digite o peso: "))
menor = peso
soma += peso

if peso > 80:
    acima80 += 1

for i in range(5):

    peso = float(input("Digite o peso: "))

    soma += peso

    if peso < menor:
        menor = peso

    if peso > 80:
        acima80 += 1

media = soma / 6

print("Média dos pesos:", media)
print("Acima de 80kg:", acima80)
print("Menor peso:", menor)
```

```python
# 09 - Votação Eletrônica

v1 = 0
v2 = 0
v3 = 0

voto = int(input("Digite o voto (0 para encerrar): "))

while voto != 0:

    if voto == 1:
        v1 += 1
    elif voto == 2:
        v2 += 1
    elif voto == 3:
        v3 += 1

    voto = int(input("Digite o voto (0 para encerrar): "))

print("Candidato 1:", v1)
print("Candidato 2:", v2)
print("Candidato 3:", v3)

if v1 > v2 and v1 > v3:
    print("Vencedor: Candidato 1")
elif v2 > v1 and v2 > v3:
    print("Vencedor: Candidato 2")
else:
    print("Vencedor: Candidato 3")
```

```python
# 10 - Produção Semanal

soma = 0
maior = 0
menos100 = 0

for i in range(7):

    producao = int(input("Digite a produção do dia: "))

    soma += producao

    if producao > maior:
        maior = producao

    if producao < 100:
        menos100 += 1

media = soma / 7

print("Média:", media)
print("Maior produção:", maior)
print("Dias abaixo de 100:", menos100)
```

```python
# 11 - Múltiplos de Três

qtd = 0
soma = 0

num = int(input("Digite um número (-1 para sair): "))

while num != -1:

    if num % 3 == 0:
        qtd += 1
        soma += num

    num = int(input("Digite um número (-1 para sair): "))

if qtd > 0:
    media = soma / qtd
    print("Quantidade:", qtd)
    print("Média:", media)
else:
    print("Nenhum múltiplo de 3.")
```

```python
# 12 - Altura dos Alunos

alturas = []
soma = 0

for i in range(5):

    altura = float(input("Digite a altura: "))
    alturas.append(altura)
    soma += altura

media = soma / 5

acima = 0

for altura in alturas:
    if altura > media:
        acima += 1

print("Média:", media)
print("Acima da média:", acima)
```

```python
# 13 - Registro de Depósitos

total = 0
qtd = 0
maior = 0

deposito = float(input("Digite o depósito (0 para sair): "))

while deposito != 0:

    total += deposito
    qtd += 1

    if deposito > maior:
        maior = deposito

    deposito = float(input("Digite o depósito (0 para sair): "))

print("Total depositado:", total)
print("Maior depósito:", maior)
print("Quantidade:", qtd)
```

```python
# 14 - Verificador de Número Primo

num = int(input("Digite um número: "))

divisores = 0

for i in range(1, num + 1):

    if num % i == 0:
        print(i)
        divisores += 1

print("Quantidade de divisores:", divisores)

if divisores == 2:
    print("É primo")
else:
    print("Não é primo")
```

```python
# 15 - Loja de Eletrônicos

total = 0

preco = float(input("Digite o preço (0 para sair): "))

while preco != 0:

    if preco > 200:
        preco = preco * 0.95

    total += preco

    preco = float(input("Digite o preço (0 para sair): "))

print("Total a pagar:", total)
```

```python
# 16 - Tempo de Atendimento

soma = 0
clientes = 0

tempo = float(input("Digite o tempo (0 para sair): "))

while tempo != 0:

    soma += tempo
    clientes += 1

    tempo = float(input("Digite o tempo (0 para sair): "))

if clientes > 0:
    media = soma / clientes

    print("Clientes atendidos:", clientes)
    print("Tempo médio:", media)
```

```python
# 17 - Análise de 10 Números

negativos = 0
soma_positivos = 0

num = int(input("Digite um número: "))
maior = num

if num < 0:
    negativos += 1
else:
    soma_positivos += num

for i in range(9):

    num = int(input("Digite um número: "))

    if num < 0:
        negativos += 1
    else:
        soma_positivos += num

    if num > maior:
        maior = num

print("Negativos:", negativos)
print("Soma positivos:", soma_positivos)
print("Maior número:", maior)
```

```python
# 18 - Consumo de Energia

total = 0

for i in range(6):

    consumo = float(input("Digite o consumo em kWh: "))

    if consumo > 100:
        valor = consumo * 0.80
    else:
        valor = consumo * 0.50

    total += valor

print("Total arrecadado:", total)
```

```python
# 19 - Adivinhe o Número Secreto

secreto = 7
tentativas = 0

palpite = int(input("Digite um número: "))

while palpite != secreto:

    tentativas += 1

    if palpite > secreto:
        print("O número secreto é menor.")
    else:
        print("O número secreto é maior.")

    palpite = int(input("Digite um número: "))

tentativas += 1

print("Acertou!")
print("Tentativas:", tentativas)
```

```python
# 20 - Estatísticas de Notas

soma = 0
qtd = 0
maior = 0
notas = []

nota = float(input("Digite a nota (-1 para sair): "))

while nota != -1:

    notas.append(nota)
    soma += nota
    qtd += 1

    if nota > maior:
        maior = nota

    nota = float(input("Digite a nota (-1 para sair): "))

media = soma / qtd

abaixo = 0

for nota in notas:
    if nota < media:
        abaixo += 1

print("Média:", media)
print("Maior nota:", maior)
print("Notas abaixo da média:", abaixo)
```

```python
# 21 - Pesquisa Eleitoral

v1 = 0
v2 = 0
v3 = 0
validos = 0
invalidos = 0

voto = int(input("Digite o voto (0 para encerrar): "))

while voto != 0:

    if voto == 1:
        v1 += 1
        validos += 1

    elif voto == 2:
        v2 += 1
        validos += 1

    elif voto == 3:
        v3 += 1
        validos += 1

    else:
        invalidos += 1

    voto = int(input("Digite o voto (0 para encerrar): "))

print("Votos válidos:", validos)
print("Votos inválidos:", invalidos)

if v1 > v2 and v1 > v3:
    print("Vencedor: Candidato 1")

elif v2 > v1 and v2 > v3:
    print("Vencedor: Candidato 2")

else:
    print("Vencedor: Candidato 3")
```

```python
# 22 - Estacionamento

total = 0

tempo = int(input("Digite o tempo em horas (0 para sair): "))

while tempo != 0:

    valor = tempo * 10

    if tempo > 5:
        valor = valor * 0.90

    total += valor

    tempo = int(input("Digite o tempo em horas (0 para sair): "))

print("Total arrecadado:", total)
```

```python
# 23 - Divisíveis por 2 e por 3

qtd = 0
soma = 0
total = 0

num = int(input("Digite um número (0 para sair): "))

while num != 0:

    soma += num
    total += 1

    if num % 2 == 0 and num % 3 == 0:
        qtd += 1

    num = int(input("Digite um número (0 para sair): "))

if total > 0:
    media = soma / total

    print("Divisíveis por 2 e 3:", qtd)
    print("Média:", media)
```

```python
# 24 - Classificação Escolar

aprovados = 0
recuperacao = 0
reprovados = 0

for i in range(8):

    nota = float(input("Digite a média do aluno: "))

    if nota >= 7:
        aprovados += 1

    elif nota >= 5:
        recuperacao += 1

    else:
        reprovados += 1

print("Aprovados:", aprovados)
print("Recuperação:", recuperacao)
print("Reprovados:", reprovados)
```

```python
# 25 - Análise de Vendas

vendas = []
soma = 0
qtd = 0

valor = float(input("Digite o valor da venda (0 para sair): "))

while valor != 0:

    vendas.append(valor)
    soma += valor
    qtd += 1

    valor = float(input("Digite o valor da venda (0 para sair): "))

if qtd > 0:

    media = soma / qtd
    acima = 0

    for venda in vendas:

        if venda > media:
            acima += 1

    print("Média das vendas:", media)
    print("Vendas acima da média:", acima)
```

```python
# 26 - Perfil Etário

soma = 0
qtd = 0
maiores = 0

idade = int(input("Digite a idade (0 para sair): "))

while idade != 0:

    soma += idade
    qtd += 1

    if idade > 18:
        maiores += 1

    idade = int(input("Digite a idade (0 para sair): "))

if qtd > 0:

    media = soma / qtd

    print("Média das idades:", media)

print("Maiores de idade:", maiores)
```

```python
# 27 - Par ou Ímpar até N

n = int(input("Digite um número: "))

for i in range(1, n + 1):

    if i % 2 == 0:
        print(i, "- Par")

    else:
        print(i, "- Ímpar")
```

```python
# 28 - Caça aos Primos

qtd_primos = 0
soma_nao_primos = 0

num = int(input("Digite um número (-1 para sair): "))

while num != -1:

    divisores = 0

    for i in range(1, num + 1):

        if num % i == 0:
            divisores += 1

    if divisores == 2:
        qtd_primos += 1

    else:
        soma_nao_primos += num

    num = int(input("Digite um número (-1 para sair): "))

print("Quantidade de primos:", qtd_primos)
print("Soma dos não primos:", soma_nao_primos)
```

```python
# 29 - Avaliação de Notas

notas = []
soma = 0
qtd = 0

nota = float(input("Digite a nota (0 para sair): "))

while nota != 0:

    notas.append(nota)
    soma += nota
    qtd += 1

    nota = float(input("Digite a nota (0 para sair): "))

if qtd > 0:

    media = soma / qtd
    acima = 0

    for n in notas:

        if n > media:
            acima += 1

    print("Média geral:", media)
    print("Notas acima da média:", acima)
```

```python
# 30 - Cálculo de Fatorial

num = int(input("Digite um número positivo: "))

fatorial = 1

for i in range(1, num + 1):

    fatorial *= i

print("Fatorial:", fatorial)

if fatorial % 2 == 0:
    print("O fatorial é par")

else:
    print("O fatorial é ímpar")
```
