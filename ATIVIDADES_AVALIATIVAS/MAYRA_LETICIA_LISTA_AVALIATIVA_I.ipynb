```python
#1- Classificador de três números
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))
c = int(input("Digite o terceiro número: "))

# Maior
if a >= b and a >= c:
    maior = a
elif b >= a and b >= c:
    maior = b
else:
    maior = c

# Menor
if a <= b and a <= c:
    menor = a
elif b <= a and b <= c:
    menor = b
else:
    menor = c

# Meio
if (a != maior and a != menor) or (a == b or a == c):
    meio = a
elif (b != maior and b != menor) or (b == a or b == c):
    meio = b
else:
    meio = c

print(f"Maior: {maior}, Meio: {meio}, Menor: {menor}")
```

```python
#2- Calculadora de Gorjeta
valor = float(input("Valor da conta: "))
nota = int(input("Nota (1 a 5): "))

if valor <= 0 or nota < 1 or nota > 5:
    print("Entrada inválida")
else:
    if nota == 5:
        gorjeta = valor * 0.20
    elif nota == 4:
        gorjeta = valor * 0.15
    elif nota == 3:
        gorjeta = valor * 0.10
    else:
        gorjeta = 0

    total = valor + gorjeta
    print(f"Gorjeta: R${gorjeta:.2f}")
    print(f"Total: R${total:.2f}")
```

```python
#3- Verificador de Ano bissexto
ano = int(input("Digite o ano: "))

if ano <= 0:
    print("Ano inválido")
else:
    if (ano % 400 == 0):
        print("Bissexto")
    else:
        if (ano % 100 == 0):
            print("Não bissexto")
        else:
            if (ano % 4 == 0):
                print("Bissexto")
            else:
                print("Não bissexto")
```

```python
#4- Cálculo de imposto de renda simplificado
salario = float(input("Salário bruto: "))

if salario <= 0:
    print("Valor inválido")
else:
    if salario <= 2259:
        aliquota = 0
    elif salario <= 2826:
        aliquota = 0.075
    elif salario <= 3751:
        aliquota = 0.15
    elif salario <= 4664:
        aliquota = 0.225
    else:
        aliquota = 0.275

    imposto = salario * aliquota
    liquido = salario - imposto

    print(f"Alíquota: {aliquota*100:.1f}%")
    print(f"Imposto: R${imposto:.2f}")
    print(f"Salário líquido: R${liquido:.2f}")
```

```python
#6- Validador de data
dia = int(input("Dia: "))
mes = int(input("Mês: "))
ano = int(input("Ano: "))

if ano <= 0:
    print("Ano inválido")
elif mes < 1 or mes > 12:
    print("Mês inválido")
else:
    bissexto = False
    if (ano % 400 == 0):
        bissexto = True
    elif (ano % 100 == 0):
        bissexto = False
    elif (ano % 4 == 0):
        bissexto = True

    if mes in (4, 6, 9, 11):
        max_dia = 30
    elif mes == 2:
        if bissexto:
            max_dia = 29
        else:
            max_dia = 28
    else:
        max_dia = 31

    if dia < 1 or dia > max_dia:
        print("Dia inválido")
    else:
        print("Data válida")
```

```python
#8- Simulador de caixa eletrônico
saldo = 1000
senha_correta = 1234

valor = float(input("Valor do saque: "))
senha = int(input("Senha: "))

if valor <= 0:
    print("Valor inválido")
else:
    if senha != senha_correta:
        print("Senha incorreta")
    else:
        if valor > saldo:
            print("Saldo insuficiente")
        else:
            saldo -= valor
            print(f"Saque realizado. Saldo: R${saldo:.2f}")
```

```python
#9- Conversor de temperatura completo
temp = float(input("Temperatura: "))
unidade = input("Unidade (C/F/K): ")

match unidade:
    case "C":
        temp_c = temp
        temp_f = (temp * 9/5) + 32
        temp_k = temp + 273.15
    case "F":
        temp_c = (temp - 32) * 5/9
        temp_f = temp
        temp_k = temp_c + 273.15
    case "K":
        if temp < 0:
            print("Kelvin inválido")
        else:
            temp_c = temp - 273.15
            temp_f = (temp_c * 9/5) + 32
            temp_k = temp
    case _:
        print("Unidade inválida")

print(f"C: {temp_c:.2f}, F: {temp_f:.2f}, K: {temp_k:.2f}")
```

```python
#10- Classificador de IMC com risco
peso = float(input("Peso: "))
altura = float(input("Altura: "))

if peso <= 0 or altura < 0.5 or altura > 2.5:
    print("Valores inválidos")
else:
    imc = peso / (altura ** 2)

    if imc < 18.5:
        classe = "Abaixo do peso"
        risco = "baixo"
    elif imc < 25:
        classe = "Normal"
        risco = "baixo"
    elif imc < 30:
        classe = "Sobrepeso"
        risco = "moderado"
    elif imc < 35:
        classe = "Obesidade I"
        risco = "moderado"
    elif imc < 40:
        classe = "Obesidade II"
        risco = "alto"
    else:
        classe = "Obesidade III"
        risco = "alto"

    print(f"IMC: {imc:.2f}")
    print(f"Classificação: {classe}")
    print(f"Risco: {risco}")
```

```python
#12- Calculadora de desconto progressivo
preco = float(input("Preço: "))
cupom = input("Cupom: ")

match cupom:
    case "PROMO10":
        desconto = 0.10
    case "PROMO20":
        desconto = 0.20
    case "PROMO30":
        desconto = 0.30
    case "SEMCUPOM":
        desconto = 0
    case _:
        print("Cupom inválido")
        desconto = 0

if preco < 50:
    if desconto > 0.10:
        desconto = 0.10

valor_desc = preco * desconto
final = preco - valor_desc

print(f"Desconto: R${valor_desc:.2f}")
print(f"Final: R${final:.2f}")
```

```python
#13- Avaliador de triângulo completo
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

if a <= 0 or b <= 0 or c <= 0:
    print("Lados inválidos")
else:
    if a < b + c and b < a + c and c < a + b:
        if a == b and b == c:
            print("Equilátero")
        elif a == b or a == c or b == c:
            print("Isósceles")
        else:
            print("Escaleno")
    else:
        print("Não forma triângulo")
```

```python
#14- calculadora de notas com conceito
n1 = float(input("Nota 1: "))
n2 = float(input("Nota 2: "))
n3 = float(input("Nota 3: "))
n4 = float(input("Nota 4: "))

if (0 <= n1 <= 10 and 0 <= n2 <= 10 and
    0 <= n3 <= 10 and 0 <= n4 <= 10):

    media = (n1 + n2 + n3 + n4) / 4

    if media >= 9:
        conceito = "A"
        status = "Aprovado"
    elif media >= 7:
        conceito = "B"
        status = "Aprovado"
    elif media >= 5:
        conceito = "C"
        status = "Recuperação"
    elif media >= 3:
        conceito = "D"
        status = "Reprovado"
    else:
        conceito = "E"
        status = "Reprovado"

    print(f"Média: {media:.2f}")
    print(f"Conceito: {conceito}")
    print(f"Status: {status}")
else:
    print("Notas inválidas")
```

```python
#15- Verificador de acesso por horário
hora = int(input("Digite a hora atual (0-23): "))
tipo = input("Digite o tipo de usuário (admin, operador, visitante): ").lower()

# Validação da hora
if hora < 0 or hora > 23:
    print("Hora inválida!")
else:
    # Determinar turno
    if 6 <= hora <= 11:
        turno = "manhã"
    elif 12 <= hora <= 17:
        turno = "tarde"
    elif 18 <= hora <= 23:
        turno = "noite"
    else:
        turno = "madrugada"

    acesso = "negado"

    match tipo:
        case "admin":
            acesso = "permitido"

        case "operador":
            if turno == "manhã" or turno == "tarde":
                acesso = "permitido"
            else:
                acesso = "negado"

        case "visitante":
            if turno == "tarde":
                acesso = "permitido"
            else:
                acesso = "negado"

        case _:
            print("Tipo de usuário inválido!")
            exit()

    print(f"Acesso {acesso}. Turno atual: {turno}.")
```

```python
# 16 - Simulador de Bomba de Combustível


gasolina = 6.75
etanol = 1.99
diesel = 6.20

tipo = input("Digite o combustível (gasolina, etanol, diesel): ").lower()
quantidade = float(input("Digite a quantidade de litros: "))


match tipo:
    case "gasolina":
        preco = gasolina
    case "etanol":
        preco = etanol
    case "diesel":
        preco = diesel
    case _:
        print("Combustível inválido!")
        preco = None

if preco is not None:
    if quantidade <= 0:
        print("Quantidade inválida!")
    else:
        subtotal = quantidade * preco
        if quantidade > 40:
            desconto = subtotal * 0.03
        else:
            desconto = 0
        total = subtotal - desconto

        print(f"Subtotal: R${(subtotal):.2f}\nDesconto: R${desconto:.2f}\nTotal: R${total:.2f}")
```

```python
# 19 - Detector de Intervalo com Tolerância

alvo=float(input("Digite um valor alvo: "))
tolerancia=float(input("Digite a tolerância: "))
medido=float(input("Digite o valor medido: "))

if alvo >= 0 and tolerancia >= 0 and medido >= 0:
    limite_inferior = alvo - tolerancia
    limite_superior = alvo + tolerancia
    if medido < limite_inferior:
        print(f'Alvo: {alvo} - Desvio: {alvo-medido} - Classificação: Abaixo do limite')
    elif limite_inferior <= medido <= limite_superior :
        print(f'Alvo: {alvo} - Desvio: {alvo-medido} - Classificação: Dentro da faixa')
    else:
        print(f'Alvo: {alvo} - Desvio: {medido-alvo} - Classificação: Acima do limite')
```

```python
# 22 - Verificador de Intervalo Múltiplo

A = int(input('Digite um intervalo mínimo: '))
B = int(input('Digite um intervalo máximo: '))
N = int(input('Digite um número: '))
D = int(input('Digite um divisor (positivo e não nulo): '))


if A >= B:
    print('Intervalo inválido: A deve ser menor que B.')
elif D <= 0:
    print('Divisor inválido: deve ser positivo e não nulo.')
else:
    if A <= N <= B:
        if N % D == 0:
            print(f'O número {N} está dentro do intervalo [{A},{B}] e é divisível por {D}')
        else:
            print(f'O número {N} está dentro do intervalo [{A},{B}] e não é divisível por {D}')
    else:
        if N % D == 0:
            print(f'O número {N} está fora do intervalo [{A},{B}] e é divisível por {D}')
        else:
            print(f'O número {N} está fora do intervalo [{A},{B}] e não é divisível por {D}')
```

```python
# 23 - Sistema de Pontuação de Jogo

jogador_1 = int(input('Digite a pontuação do 1º jogador: '))
jogador_2 = int(input('Digite a pontuação do 2º jogador: '))
jogador_3 = int(input('Digite a pontuação do 3º jogador: '))

if 0 <= jogador_1 <= 1000 and 0 <= jogador_2 <= 1000 and 0 <= jogador_3 <= 1000:
    if jogador_1 == jogador_2 == jogador_3:
        print('Todos os jogadores empataram com a mesma pontuação.')
    else:
        if jogador_1 >= jogador_2 and jogador_1 >= jogador_3:
            primeiro = jogador_1
            if jogador_2 >= jogador_3:
                segundo = jogador_2
                terceiro = jogador_3
            else:
                segundo = jogador_3
                terceiro = jogador_2
        elif jogador_2 >= jogador_1 and jogador_2 >= jogador_3:
            primeiro = jogador_2
            if jogador_1 >= jogador_3:
                segundo = jogador_1
                terceiro = jogador_3
            else:
                segundo = jogador_3
                terceiro = jogador_1
        else:
            primeiro = jogador_3
            if jogador_1 >= jogador_2:
                segundo = jogador_1
                terceiro = jogador_2
            else:
                segundo = jogador_2
                terceiro = jogador_1

        diferenca = primeiro - terceiro

        print(f'\nRANKING FINAL:\n1º lugar: {primeiro}\n2º lugar: {segundo}\n3º lugar: {terceiro}\nDiferença entre 1º e 3º: {diferenca}')
else:
    print('Pontuação inválida. Digite valores entre 0 e 1000.')
```

```python
# 24 - Calculadora de Reajuste Salarial

salario = float(input("Digite o salario atual: "))
desempenho = int(input("Digite a avaliação de desempenho: "))

if salario >= 0 and 0 <= desempenho <= 100:
    if desempenho >= 90:
        reajuste = salario * 1.15
        percentual = 15
    elif desempenho >= 75:
        reajuste = salario * 1.10
        percentual = 10
    elif desempenho >= 60:
        reajuste = salario * 1.05
        percentual = 5
    else:
        reajuste = salario
        percentual = 0
    if reajuste > 15000:
        print(f'O salário atingiu o teto máximo de R$15.000,00 - Salário anterior: R${salario:.2f} - Percentual aplicado: {percentual}% - Reajuste: R${15000 - salario:.2f} - Novo salário: R$15000.00')
    else:
        print(f'Salário anterior: R${salario:.2f} -Percentual aplicado: {percentual}% - Reajuste: R${reajuste - salario:.2f} - Novo salário: R${reajuste:.2f}')
else:
    print('Valores inválidos.')
```

```python
# 25 - Classificador de Triângulo com Ângulos

angulo1 = int(input('Digite o 1º angulo do triângulo em graus: '))
angulo2 = int(input('Digite o 2º angulo do triângulo em graus: '))
angulo3 = int(input('Digite o 3º angulo do triângulo em graus: '))

if angulo1 >= 0 and angulo2 >= 0 and angulo3 >= 0:
    if angulo1 + angulo2 + angulo3 != 180:
        print(f"Trinângulo inválido. Soma dos angulos = {angulo1+angulo2+angulo3}°")
    elif angulo1 < 90 and angulo2 < 90 and angulo3 < 90:
        print(f"Trinângulo acutângulo (todos os angulos < 90°)")
        if angulo1 == angulo2 == angulo3:
            print("Classificação quanto aos lados: Equilátero")
        elif angulo1 == angulo2 or angulo1 == angulo3 or angulo2 == angulo3:
            print("Classificação quanto aos lados: Isósceles")
        else:
            print("Classificação quanto aos lados: Escaleno")
    elif angulo1 == 90 or angulo2 == 90 or angulo3 == 90:
        print(f"Trinângulo retângulo (um angulo é = 90°)")
        if angulo1 == angulo2 == angulo3:
            print("Classificação quanto aos lados: Equilátero")
        elif angulo1 == angulo2 or angulo1 == angulo3 or angulo2 == angulo3:
            print("Classificação quanto aos lados:  Isósceles")
        else:
            print("Classificação quanto aos lados: Escaleno")
    else:
        print(f"Trinângulo obtusângulo (um angulo é > 90°)")
        if angulo1 == angulo2 == angulo3:
            print("Classificação quanto aos lados: Equilátero")
        elif angulo1 == angulo2 or angulo1 == angulo3 or angulo2 == angulo3:
            print("Classificação quanto aos lados: Isósceles")
        else:
            print("Classificação quanto aos lados: Escaleno")
```

```python
# 26 - Calculadora de Combustível de Viagem

distancia = float(input('Digite a distância percorrida (km): '))
consumo = float(input('Digite o consumo do veículo (km/l): '))
preco = float(input('Digite o preço do litro de combustível (R$/l): '))

if distancia > 0 and consumo > 0 and preco > 0:
    litros = distancia/consumo
    if litros * preco > 500:
        print(f'Para o percurso de {distancia:.2f}km serão necessários {litros:.2f}l de combustível - Valor total: R${preco*litros:.2f}. Parada sugerida em {distancia/2:.2f}km para reabastecimento')
    else:
        print(f'Para o percurso de {distancia:.2f}km serão necessários {litros:.2f}l de combustível - Valor total: R${preco*litros:.2f}')
else:
    print('Valores inválidos. Digite apenas números positivos.')
```

```python
# 27 - Verificador de Senha com Nível de Segurança

senha = input('Digite a senha: ')
confirmacao = input('Digite novamente a senha: ')

if senha != confirmacao:
    print('Senhas não conferem')
else:
    if 0 <= len(senha) < 6:
        nivel = 'Fraca'
    elif 6 <= len(senha) <= 9:
        nivel = 'Média'
    else:
        nivel = 'Forte'
    if senha.isdigit():
        if nivel == 'Forte':
            nivel = 'Média'
        elif nivel == 'Média':
            nivel = 'Fraca'
    print(nivel)
```

```python
# 28 - Simulador de Reserva de Assento

assento = int(input('Digite o número do assento: '))
classe = input('Digite a  classe (economica, executiva ou primeira): ').lower()

if 1 <= assento <= 100:
    match classe:
        case "primeira":
            if assento <= 10:
                print('Número do assento e classe compatíveis. Valor: R$1.500,00. Deseja reservar?')
            else:
                print("O número do assento não é compatível com a classe.")
        case "executiva":
            if 11 <= assento <= 30:
                print('Número do assento e classe compatíveis. Valor: R$890,00. Deseja reservar?')
            else:
                print("O número do assento não é compatível com a classe.")
        case "economica":
            if 31 <= assento <= 100:
                print('Número do assento e classe compatíveis. Valor: R$350,00. Deseja reservar?')
            else:
                print("O número do assento não é compatível com a classe.")
        case _:
            print("Classe inválida.")
else:
    print("Número de assento inválido.")
```

```python
# 29 - Simulador de Batalha por Tipos

atacante = input('Digite o tipo do atacante (fogo, agua, planta, eletrico, terra): ').lower()
defensor = input('Digite o tipo do defensor (fogo, agua, planta, eletrico, terra): ').lower()
poder = float(input('Digite o poder base do ataque: '))

match atacante:
    case 'fogo':
        if defensor == 'planta':
            dano_final = poder * 2
        elif defensor == 'agua':
            dano_final = poder * 0.5
        else:
            dano_final = poder
    case 'agua':
        if defensor == 'fogo':
            dano_final = poder * 2
        elif defensor == 'planta':
            dano_final = poder * 0.5
        else:
            dano_final = poder
    case 'planta':
        if defensor == 'agua':
            dano_final = poder * 2
        elif defensor == 'fogo':
            dano_final = poder * 0.5
        else:
            dano_final = poder
    case 'eletrico':
        if defensor == 'agua':
            dano_final = poder * 2
        elif defensor == 'terra':
            dano_final = poder * 0.5
        else:
            dano_final = poder
    case 'terra':
        if defensor == 'eletrico':
            dano_final = poder * 2
        elif defensor == 'agua':
            dano_final = poder * 0.5
        else:
            dano_final = poder
    case _:
        print('Tipo inválido.')

print(f'Dano final causado: {dano_final:.2f}')
```

```python
# 30 - Sistema Completo de Reserva de Hotel

tipo = input('Digite o tipo de acomodação (standard, luxo, suite): ').lower()
diaria = int(input('Digite o número de diárias: '))
fidelidade = input('Possui cartão fidelidade(s/n)? ').lower()

standard = 280.00
suite = 450.00
luxo = 800.00

if diaria > 0:
    match tipo:
        case 'standard':
            diaria_base = standard * diaria
        case 'luxo':
            diaria_base = luxo * diaria
        case 'suite':
            diaria_base = suite * diaria
        case _:
            print('Tipo de acomodação inválido.')
    if 1 <= diaria <= 3:
        desconto_qtd = 0
    elif 4 <= diaria <= 7:
        desconto_qtd = diaria_base * 0.10
    else:
        desconto_qtd = diaria_base * 0.20

    subtotal = diaria_base - desconto_qtd

    if fidelidade == 's':
        desconto_fidelidade = subtotal * 0.05
    else:
        desconto_fidelidade = 0

    total_final = subtotal - desconto_fidelidade

    if total_final < 100:
        total_final = 100

    print(f'Resumo da reserva:\nTipo: {tipo}\nDiárias: {diaria}\nValor base: R${diaria_base:.2f}\nDesconto por quantidade: R${desconto_qtd:.2f}\nDesconto fidelidade: R${desconto_fidelidade:.2f}\nTotal final: R${total_final:.2f}')
else:
    print('Número de diárias inválido.')
```
