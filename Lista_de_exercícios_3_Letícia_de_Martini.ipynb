```python
17
matriz = []

for i in range(3):
    linha = []

    for j in range(3):
        valor = float(input("Digite um valor: "))
        linha.append(valor)

    matriz.append(linha)

det = (
    matriz[0][0] * matriz[1][1] * matriz[2][2]
    + matriz[0][1] * matriz[1][2] * matriz[2][0]
    + matriz[0][2] * matriz[1][0] * matriz[2][1]
    - matriz[0][2] * matriz[1][1] * matriz[2][0]
    - matriz[0][0] * matriz[1][2] * matriz[2][1]
    - matriz[0][1] * matriz[1][0] * matriz[2][2]
)

print("Determinante =", det)
```

```python
18
matriz1 = []
matriz2 = []

print("Primeira matriz")

for i in range(3):
    linha = []

    for j in range(3):
        valor = int(input("Digite um valor: "))
        linha.append(valor)

    matriz1.append(linha)

print("Segunda matriz")

for i in range(3):
    linha = []

    for j in range(3):
        valor = int(input("Digite um valor: "))
        linha.append(valor)

    matriz2.append(linha)

soma = []
diferenca = []

for i in range(3):
    linha_soma = []
    linha_dif = []

    for j in range(3):
        linha_soma.append(matriz1[i][j] + matriz2[i][j])
        linha_dif.append(matriz1[i][j] - matriz2[i][j])

    soma.append(linha_soma)
    diferenca.append(linha_dif)

print("Matriz Soma:")
for linha in soma:
    print(linha)

print("Matriz Diferença:")
for linha in diferenca:
    print(linha)
```

```python
23
A = set()
B = set()

tamA = int(input("Quantidade de elementos do conjunto A: "))

for i in range(tamA):
    num = int(input("Digite um número: "))
    A.add(num)

tamB = int(input("Quantidade de elementos do conjunto B: "))

for i in range(tamB):
    num = int(input("Digite um número: "))
    B.add(num)

if A.issubset(B):
    print("A é subconjunto de B")
elif B.issubset(A):
    print("B é subconjunto de A")
else:
    print("Nenhum conjunto é subconjunto do outro")
```

```python
24
conjunto = set()

quantidade = int(input("Quantos números deseja digitar? "))

for i in range(quantidade):
    num = int(input("Digite um número: "))
    conjunto.add(num)

quadrados = set()

for num in conjunto:
    i = 0

    while i * i <= num:
        if i * i == num:
            quadrados.add(num)

        i += 1

print("Quadrados perfeitos:", quadrados)
```

```python
29
produtos = {}

for i in range(3):
    nome = input("Digite o nome do produto: ")

    precos = []

    for j in range(3):
        preco = float(input("Digite o preço: "))
        precos.append(preco)

    produtos[nome] = precos

maior_variacao = 0
produto_maior = ""

for produto in produtos:
    variacao = max(produtos[produto]) - min(produtos[produto])

    if variacao > maior_variacao:
        maior_variacao = variacao
        produto_maior = produto

print("Produto com maior variação:", produto_maior)
print("Variação:", maior_variacao)
```
